```python
import csv
import boto3

# Initialize DynamoDB resource
dynamodb = boto3.resource('dynamodb', region_name='us-west-2', endpoint_url='http://172.20.0.10:8000',
                          aws_access_key_id='local', aws_secret_access_key='local')

# Define table schema
table_name = 'Car_Catalog'
table_schema = {
    'TableName': table_name,
   'KeySchema': [
        {'AttributeName': 'car_id', 'KeyType': 'HASH'},  # Partition key (using UUID)
    ],
    'AttributeDefinitions': [
        {'AttributeName': 'car_id', 'AttributeType': 'S'},  # S for string
        # Add more attribute definitions as needed
    ],
    'BillingMode': 'PAY_PER_REQUEST'  # You can change this to PROVISIONED if needed
}

# Create DynamoDB table
table = dynamodb.create_table(**table_schema)

# Wait for table to be created (optional)
table.wait_until_exists()

# Function to read CSV file and return a list of dictionaries
def read_csv(filename):
    data = []
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(row)
    return data

# Function to insert data into DynamoDB table
def insert_data(table, data):
    with table.batch_writer() as batch:
        for item in data:
            batch.put_item(Item=item)

# Example usage
if __name__ == "__main__":
    csv_filename = './Cleaned/Immatriculations/part-00000-6e7cc082-8fb0-4819-8fe8-cf3f68aa8a5b-c000.csv'
    csv_data = read_csv(csv_filename)
    insert_data(table, csv_data)
```

```python
import csv
import boto3
import uuid

# Initialize DynamoDB resource
dynamodb = boto3.resource('dynamodb', region_name='us-west-2', endpoint_url='http://172.20.0.11:8000',
                          aws_access_key_id='local', aws_secret_access_key='local')

# Define table schema
table_name = 'CarCatalog'
table_schema = {
    'TableName': table_name,
    'KeySchema': [
        {'AttributeName': 'car_id', 'KeyType': 'HASH'},  # Partition key (using UUID)
    ],
    'AttributeDefinitions': [
        {'AttributeName': 'car_id', 'AttributeType': 'S'},  # S for string
        # Add more attribute definitions as needed
    ],
    'BillingMode': 'PAY_PER_REQUEST'  # You can change this to PROVISIONED if needed
}

# Create DynamoDB table
table = dynamodb.create_table(**table_schema)

# Wait for table to be created (optional)
table.wait_until_exists()

# Function to read CSV file and return a list of dictionaries
def read_csv(filename):
    data = []
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            # Generate UUID for primary key
            row['car_id'] = str(uuid.uuid4())
            data.append(row)
    return data

# Function to insert data into DynamoDB table
def insert_data(table, data):
    with table.batch_writer() as batch:
        for item in data:
            batch.put_item(Item=item)

# Example usage
if __name__ == "__main__":
    csv_filename = './Cleaned/catalogue/part-00000-95f65e95-e35e-4813-8788-69102107ae0c-c000.csv'
    csv_data = read_csv(csv_filename)
    insert_data(table, csv_data)
```
